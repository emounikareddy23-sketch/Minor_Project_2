```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("Data set for DADS June(2).csv")

print(df.head())
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print("Duplicate Records :", duplicates)

df = df.drop_duplicates()

print("Shape After Removing Duplicates :", df.shape)
```

```python
df['Date'] = pd.to_datetime(
    df['Date'],
    errors='coerce',
    dayfirst=True
)

# Remove invalid dates
df = df.dropna(subset=['Date'])

# Create new date columns
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
df['Month_Number'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Day_Name'] = df['Date'].dt.day_name()

print(df[['Date','Month','Day_Name']].head())
```

```python
df['Amount'] = (
    df['Amount']
    .astype(str)
    .str.replace('₹','',regex=False)
    .str.replace('Rs.','',regex=False)
    .str.replace(',','',regex=False)
    .str.strip()
)

df['Amount'] = pd.to_numeric(df['Amount'],errors='coerce')

df = df.dropna(subset=['Amount'])

print(df['Amount'].head())
```

```python
df['Type'] = df['Type'].replace({
    'DR':'debit',
    'Debit':'debit',
    'CR':'credit',
    'Credit':'credit'
})

print(df['Type'].value_counts())
```

```python
vendor_dict = {

'Swiggy':['SWIGGY','BUNDL'],

'Zomato':['ZOMATO'],

'Amazon':['AMAZON','AMZN','AMAZONPAY'],

'Flipkart':['FLIPKART'],

'Myntra':['MYNTRA'],

'Zepto':['ZEPTO'],

'Blinkit':['BLINKIT','GROFERS'],

'BigBasket':['BIGBASKET'],

'DMart':['DMART'],

'Uber':['UBER'],

'Ola':['OLA'],

'Rapido':['RAPIDO'],

'IRCTC':['IRCTC'],

'BookMyShow':['BOOKMYSHOW'],

'Netflix':['NETFLIX'],

'Spotify':['SPOTIFY'],

'Hotstar':['HOTSTAR','DISNEY'],

'Prime':['PRIME'],

'Youtube':['YOUTUBE'],

'Jio':['JIO'],

'Airtel':['AIRTEL'],

'BSNL':['BSNL'],

'ACT':['ACT'],

'Electricity':['BESCOM'],

'Water':['BWSSB'],

'Gas':['HP GAS','INDANE'],

'Zerodha':['ZERODHA','COIN'],

'Groww':['GROWW'],

'Paytm':['PAYTM'],

'PhonePe':['PHONEPE'],

'Cash Withdrawal':['ATM'],

'P2P Transfer':['ANKIT','RAHUL','PRIYA','ROHIT','VIVEK','NEHA']
}

def extract_vendor(desc):

    desc = str(desc).upper()

    for vendor,keywords in vendor_dict.items():

        for word in keywords:

            if word in desc:

                return vendor

    return "Uncategorised"

df['Vendor'] = df['Description'].apply(extract_vendor)

print(df['Vendor'].value_counts().head(20))
```

```python
category_dict = {

'Swiggy':'Food Delivery',

'Zomato':'Food Delivery',

'Zepto':'Quick Commerce',

'Blinkit':'Quick Commerce',

'BigBasket':'Groceries',

'DMart':'Groceries',

'Amazon':'E-Commerce',

'Flipkart':'E-Commerce',

'Myntra':'E-Commerce',

'Uber':'Transport',

'Ola':'Transport',

'Rapido':'Transport',

'IRCTC':'Transport',

'BookMyShow':'Entertainment',

'Netflix':'Subscriptions',

'Spotify':'Subscriptions',

'Hotstar':'Subscriptions',

'Prime':'Subscriptions',

'Youtube':'Subscriptions',

'Jio':'Utilities',

'Airtel':'Utilities',

'BSNL':'Utilities',

'ACT':'Utilities',

'Electricity':'Utilities',

'Water':'Utilities',

'Gas':'Utilities',

'Zerodha':'Investments',

'Groww':'Investments',

'Cash Withdrawal':'Cash Withdrawal',

'P2P Transfer':'Personal Transfer',

'Paytm':'Wallet',

'PhonePe':'Wallet'

}

df['Category'] = df['Vendor'].map(category_dict)

df['Category'] = df['Category'].fillna('Others')

print(df['Category'].value_counts())
```

```python
# Credit Transactions
credit_df = df[df['Type'] == 'credit']

# Debit Transactions
debit_df = df[df['Type'] == 'debit']

total_credit = credit_df['Amount'].sum()
total_debit = debit_df['Amount'].sum()

net_savings = total_credit - total_debit

savings_rate = (net_savings / total_credit) * 100

print("="*60)
print("SPENDING OVERVIEW")
print("="*60)

print("Total Credits :", round(total_credit,2))
print("Total Debits  :", round(total_debit,2))
print("Net Savings   :", round(net_savings,2))
print("Savings Rate  :", round(savings_rate,2),"%")
print("Transactions  :", len(df))
print("Unique Vendors:", df['Vendor'].nunique())

print("="*60)

# Top Categories

category_spend = debit_df.groupby('Category')['Amount'].sum()

category_spend = category_spend.sort_values(ascending=False)

print(category_spend)

print("="*60)

# Top Vendors

vendor_spend = debit_df.groupby('Vendor')['Amount'].sum()

vendor_spend = vendor_spend.sort_values(ascending=False)

print(vendor_spend.head(10))
```

```python
monthly_trend = debit_df.pivot_table(
    values='Amount',
    index='Category',
    columns='Month',
    aggfunc='sum',
    fill_value=0
)

print(monthly_trend)

print("="*60)

growth = ((monthly_trend.iloc[:,-1] - monthly_trend.iloc[:,0]) /
          monthly_trend.iloc[:,0]) * 100

growth = growth.replace([np.inf,-np.inf],0)

growth = growth.fillna(0)

print("Monthly Growth (%)")

print(growth.sort_values(ascending=False))
```

```python
# ==========================================
# STEP 12 : TIME-OF-DAY ANALYSIS
# ==========================================

# Copy only debit transactions
debit_df = df[df['Type'] == 'debit'].copy()

# Convert Time column to string and extract hour
debit_df['Hour'] = debit_df['Time'].astype(str).str[:2]

# Convert Hour to numeric
debit_df['Hour'] = pd.to_numeric(debit_df['Hour'], errors='coerce')

# Remove invalid hours
debit_df = debit_df.dropna(subset=['Hour'])

# Convert Hour to integer
debit_df['Hour'] = debit_df['Hour'].astype(int)

# Hour-wise spending
hourly_spending = debit_df.groupby('Hour')['Amount'].sum()

print("="*50)
print("HOURLY SPENDING")
print("="*50)
print(hourly_spending)

# Food Delivery Analysis
food_df = debit_df[debit_df['Category'] == 'Food Delivery']

if len(food_df) > 0:

    late_night = food_df[
        (food_df['Hour'] >= 21) | (food_df['Hour'] <= 2)
    ]

    percentage = (len(late_night) / len(food_df)) * 100

    print("\nFood Delivery Orders :", len(food_df))
    print("Late Night Orders :", len(late_night))
    print("Late Night Percentage :", round(percentage, 2), "%")

else:
    print("\nNo Food Delivery transactions found.")

# Category vs Hour
print("\nCategory vs Hour Spending")

category_hour = debit_df.groupby(
    ['Category', 'Hour']
)['Amount'].sum()

print(category_hour)
```

```python
# ==========================================
# STEP 13 : ANOMALY DETECTION
# ==========================================

# Filter debit transactions
debit_df = df[df["Type"] == "debit"].copy()

# Mean and Standard Deviation for each Category
debit_df["Mean"] = debit_df.groupby("Category")["Amount"].transform("mean")
debit_df["Std"] = debit_df.groupby("Category")["Amount"].transform("std")

# Avoid division by zero
debit_df["Std"] = debit_df["Std"].fillna(1)

# Calculate Z-Score
debit_df["Z_Score"] = (
    debit_df["Amount"] - debit_df["Mean"]
) / debit_df["Std"]

# Find anomalies
anomalies = debit_df[debit_df["Z_Score"] > 2]

# Sort anomalies
anomalies = anomalies.sort_values("Z_Score", ascending=False)

print("Top 10 Anomalies")
print(anomalies[["Date", "Vendor", "Category", "Amount", "Z_Score"]].head(10))
```

```python
print("=" * 60)
print("SPENDING ARCHETYPES")
print("=" * 60)

# Calculate percentages safely
food_percent = ((category_spend.get("Food Delivery", 0) +
                 category_spend.get("Restaurants", 0) +
                 category_spend.get("Cafe", 0)) / total_debit) * 100

quick_percent = (category_spend.get("Quick Commerce", 0) / total_debit) * 100
shop_percent = (category_spend.get("E-Commerce", 0) / total_debit) * 100
invest_percent = (category_spend.get("Investments", 0) / total_debit) * 100
transport_percent = (category_spend.get("Transport", 0) / total_debit) * 100

# Display Archetypes
if food_percent > 25:
    print("THE FOODIE")

if quick_percent > 15:
    print("THE QUICK COMMERCE JUNKIE")

if shop_percent > 15:
    print("THE SHOPAHOLIC")

if invest_percent > 15:
    print("THE INVESTOR")

if late_percent > 50:
    print("THE LATE-NIGHT SNACKER")

if transport_percent > 10:
    print("THE CAB COMMUTER")

if savings_rate < 10:
    print("THE YOLO SPENDER")

if savings_rate > 40:
    print("THE DISCIPLINED SAVER")

print("=" * 60)
```
